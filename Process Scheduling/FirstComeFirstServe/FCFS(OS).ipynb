```python
#declaring and initializing cariables
N = int(input("Enter the number of processes"))
process_list = []
completion_time_list = [0]
process_details = { 'process_name' : None,
                    'arrival_time' : None,
                    'burst_time' : None,
                    }
```

```python
#method to accept process details
def ProcessDetails():
    process_name = input('Enter Process Name: ')
    arrival_time = int(input('Enter Arrival Time: '))
    burst_time = int(input('Enter Burst Time: '))
    #initializing the process details                    
    for key in process_details.keys():
        if key == 'process_name':
            process_details[key] = process_name
        elif key == 'arrival_time':
            process_details[key] = arrival_time
        elif key == 'burst_time':
            process_details[key] = burst_time
    #Append a process into process list
    process_list.append(process_details.copy())
```

```python
#method call
for _ in range(N):
  ProcessDetails()
```

```python
5#display process list
process_list
```

```python
#sort the processes based on arrival time
sorted_process_list = sorted(process_list, key=lambda d: d['arrival_time']) 
sorted_process_list
```

```python
#list of all the burst time of the sorted process list
burst_time_list = [process['burst_time'] for process in sorted_process_list]
burst_time_list
```

```python
#calculate the completion time
for i in range(len(burst_time_list)):
  completion_time_list.append(burst_time_list[i]+completion_time_list[i])
completion_time_list=completion_time_list[1:]
completion_time_list
```

```python
#calculate the turnaround time Turnaround Time = Completion time – Arrival time 
arrival_time_list = [process['arrival_time'] for process in sorted_process_list]
turnaround_time_list = [(CT-AT) for CT, AT in zip(completion_time_list, arrival_time_list)]
turnaround_time_list
```

```python
# arrival_time_list = [process['arrival_time'] for process in sorted_process_list]
turnaround_time_list = completion_time_list
turnaround_time_list
```

```python
#calculate the waiting time Waiting Time = Turnaround time – Burst Time
arrival_time_list = [process['arrival_time'] for process in sorted_process_list]
waiting_time_list = [(TT-BT) for TT, BT in zip(turnaround_time_list, burst_time_list)]
waiting_time_list
```

```python
# Gantt Chart
gantt_chart = [0]
gantt_chart.extend(completion_time_list)
gantt_chart
```

```python
#Updating the waiting time to the corresponding process 
for process, waiting_time in zip(sorted_process_list, waiting_time_list):
    process['waiting_time'] = waiting_time
```

```python
#Updating the turnaround time to the corresponding process 
for process, turnaround_time in zip(sorted_process_list, turnaround_time_list):
    process['turn_around_time'] = turnaround_time
```

```python
#average waiting time
avg_waiting_time = sum(waiting_time_list)/N
```
